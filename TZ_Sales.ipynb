# Main Task:
### Obtain information about the product sales of the various retailer types in UTC offset. Display the average amount of sales that occurred in each time zone.

### Solution map:
   - <b>Subtask 1:</b> Store the date and time values in a single column called "MOS", denoting "Moment of Sale". Verify that the time zone values stored in "sales_data" are valid and can be manipulated with the pytz module. (Data Preparation)
   
   - <b>Subtask 2:</b> Estimate the offset of the values of "MOS" to UTC. Store them in a column called "OffsetUTC". (Data Manipulation)
   
   - <b>Subtask 3:</b> Order all sales according to a reconciled UTC-equivalent of the moment of sale and analyze the data. (Data Analysis)
   
   - <b>Subtask 4:</b> Obtain statistics and visualizations that will respond to the Main Task and further improve the analysis of your data. (Data Visualization)

## Introduction

```python
import pandas as pd 
import pytz 
from datetime import datetime 
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('Sales-products-tz-mod.csv', index_col = 'SaleID')
sale_data = data.copy()
sale_data.head()
```

```python
sale_data[['DateOfSale', 'TimeOfSale', 'TimeZone']].head()
```

## Part - 1 Data preparation 

- **Subtask 1:**

```python
date_and_time = sale_data['DateOfSale'] + ' ' + sale_data['TimeOfSale']
date_and_time
```

```python
sale_data['MOS'] = pd.to_datetime(date_and_time)
sale_data.head()
```

```python
sale_data  = sale_data.drop(['DateOfSale', 'TimeOfSale'], axis =1)
sale_data.head()
```

```python
pytz.all_timezones
```

```python
sale_data['TimeZone'].unique()
```

```python
for i in sale_data['TimeZone'].unique():
    if i in (pytz.all_timezones):
        print(True)
    else: 
        print(False)
```

```python
sale_data['TimeZone'].unique()
```

```python
sale_data.loc[:,'TimeZone'] = sale_data.loc[:,'TimeZone'].replace({'SGT':'Singapore'})
```

```python
sale_data['TimeZone'].unique()
```

```python
for i in sale_data['TimeZone'].unique():
    if i in (pytz.all_timezones):
        print(True)
    else: 
        print(False)
```

```python
sale_data[sale_data['TimeZone'] == "Singapore"]
```

```python
pd.options.display.max_rows = None
pd.options.display.max_columns = None
sale_data
```

## Part - 2 Data Manipulation 

- **Subtask 2:**

```python
sale_data['MOS'].iloc[0]
```

```python
sale_data['MOS'].iloc[0].tz_localize(sale_data['TimeZone'].iloc[0])
```

```python
list_mos_timestamps = []
list_mos_timestamps
```

```python
for i in range(len(sale_data)):
    lo = pd.to_datetime(sale_data['MOS'].iloc[i].tz_localize(
        sale_data['TimeZone'].iloc[i], ambiguous='raise', nonexistent='raise'))
    
    list_mos_timestamps.append(lo)
```

```python
list_mos_timestamps[:10]
```

```python
list_mos_timestamps
```

```python
 list_mos_timestamps[0].strftime('%Y-%m-%d %H:%M:%S')
```

```python
list_mos_timestamps[0].strftime('%z')
```

```python
float(list_mos_timestamps[0].strftime('%z'))/100
```

```python
list_mos_timestamps[0].date()
```

```python
list_mos_timestamps[0].time()
```

```python
list_mos_timestamps[0].utcoffset()
```

```python
list_mos_timestamps[0].utcoffset().total_seconds()/3600
```

```python
sale_data['OffsetUTC'] = [list_mos_timestamps[i].utcoffset().total_seconds()/3600 for i in range(len(sale_data))]
```

```python
sale_data.head()
```

```python
sale_data.columns 
```

```python
Column_name_reordered = ['RetailerCountry','MOS','TimeZone','OffsetUTC','RetailerType','Product','Sales Revenue ($)']
sale_date = sale_data[Column_name_reordered]
sale_date.head()
```

```python
sale_date.groupby(['TimeZone', 'OffsetUTC']).count()
```

## Part - 3 Data Analysis
- **Subtask 3:**

```python
sale_data.sort_values(['MOS'])
```

```python
sale_data['MOS-UTC'] = [pd.to_datetime(sale_data["MOS"].iloc[i]).tz_localize(sale_data["TimeZone"].iloc[i]).tz_convert('UTC')
                       for i in range(len(sale_data))]
```

```python
sale_data.head()
```

```python
Column_name_reordered = ['RetailerCountry','MOS-UTC','MOS','TimeZone','OffsetUTC','RetailerType','Product','Sales Revenue ($)']
sale_data = sale_data[Column_name_reordered]
sale_data.head()
```

```python
sale_data.sort_values("MOS-UTC")
```

```python
sale_data.groupby('TimeZone').mean()
```

```python
sale_data.groupby('TimeZone').count()
```

```python
sale_data.groupby(['TimeZone', 'OffsetUTC']).count()
```

```python
sale_data.groupby('TimeZone').count().sort_values('Product', ascending=False)
```

## Part - 4 Data Visualization 
- **Subtask 4:**

```python
sale_data.head()
```

```python
sales_data_viz = sale_data.reset_index()
sales_data_viz.head()
```

```python
sales_data_viz = sales_data_viz.set_index('MOS-UTC').sort_index()
sales_data_viz.head()
```

```python
plt.scatter(sales_data_viz['OffsetUTC'], sales_data_viz['RetailerType'])
```

```python
total_sales_revenue_by_timezones = sales_data_viz.groupby('TimeZone').sum()
total_sales_revenue_by_timezones
```

```python
plt.bar(total_sales_revenue_by_timezones.index,
       height = total_sales_revenue_by_timezones['Sales Revenue ($)'])
```

```python
plt.figure(figsize=(7,6))

plt.bar(total_sales_revenue_by_timezones.index,
       height = total_sales_revenue_by_timezones['Sales Revenue ($)'], color = '#108A99')

plt.xticks(rotation = 45, fontsize=13)
plt.yticks(fontsize= 13)
plt.title("Total Sales Revenue by TimeZones", fontsize=18, fontweight='bold')
plt.ylabel("Sales Revenue ($)", fontsize = 13)

plt.show()
```

```python
averages_by_timezones = sales_data_viz.groupby(['TimeZone']).mean().sort_values("Sales Revenue ($)", ascending=False)
averages_by_timezones
```

```python
plt.bar(averages_by_timezones.index,
       height= averages_by_timezones['Sales Revenue ($)'],
       color = ['green', 'blue','red','darkblue','yellow','purple'])

plt.xticks(rotation = 45, fontsize=13)
plt.yticks(fontsize= 13)
plt.title("Total Sales Revenue by TimeZones", fontsize=18, fontweight='bold')
plt.ylabel("Sales Revenue ($)", fontsize = 13)

plt.show()
```
